```python
print((data['5k']=="-").sum(),(data['10k']=="-").sum(),
    (data['20k']=="-").sum(),(data['25k']=="-").sum(),
    (data['30k']=="-").sum(),(data['40k']=="-").sum(),
    (data['half']=="-").sum())
```

# Types of Datatype

```python
count={'float64':0,'int64':0,'object':0}
for i, j in data.dtypes.items():
    count[str(j)]=count[str(j)]+1
plt.bar(range(len(count)), list(count.values()), align='center', color='orange')
plt.xticks(range(len(count)), list(count.keys()))
plt.show()
plt.savefig('images/datatype1.png')   # save the figure to file
```

# The Correlations

```python
plt.figure(figsize=(10,9))
corr = data[['division','age','official','genderdiv','overall','pace','5k','10k','25k','30k','35k','half','20k','40k']].corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, xticklabels=corr.columns.values,cmap="Dark2",
            yticklabels=corr.columns.values,mask=mask, vmax=1,
            center=0,  annot= True, square=True)
plt.savefig('images/correlations.png')   # save the figure to file
```

# Handling missing values - One Hot Encode - Normalization

```python
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
data.dropna(axis=0, how='any', inplace=True)
```

```python
dfgender= pd.get_dummies(data.gender, drop_first=True)
dfstate = pd.get_dummies(data.state, drop_first=True)
dfcountry = pd.get_dummies(data.country, drop_first=True)
dfcity =pd.get_dummies(data.city, drop_first=True)
data.drop(['country','gender','state','city'],axis=1, inplace=True)
```

```python
data  =  data.join(dfgender)
data  =  data.join(dfstate)
data  =  data.join(dfcountry)
data  =  data.join(dfcity)
```
